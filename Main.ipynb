```python
# 파일 이름, 사이즈(폭, 높이), 라벨링 오브젝트 치수(이름, x최대 최소, y최대 최소)
# filename, size(width, height), object(name, xmax, xmin, ymax, ymin)

def extract_text(filename):
    tree = et.parse(filename)
    root = tree.getroot()

    # 파일이름 가져오기
    image_name = root.find('filename').text
    # 폭, 높이 가져오기
    width = root.find('size').find('width').text
    height = root.find('size').find('height').text
    objs = root.findall('object')
    parser = []
    for obj in objs :
        name = obj.find('name').text
        bndbox = obj.find('bndbox')
        xmax = bndbox.find('xmax').text
        xmin = bndbox.find('xmin').text
        ymax = bndbox.find('ymax').text
        ymin = bndbox.find('ymin').text
        parser.append([image_name, width, height, name, xmax, xmin, ymax, ymin])

    return parser
```

```python
# step-2: read xml files
# from each xml file we need to extract
# filename, size(width, height), object(name, xmin, xmax, ymin, ymax)
def extract_text(filename):
    tree = et.parse(filename)
    root = tree.getroot()

    # extract filename
    image_name = root.find('filename').text
    # width and height of the image
    width = root.find('size').find('width').text
    height = root.find('size').find('height').text
    objs = root.findall('object')
    parser = []
    for obj in objs:
        name = obj.find('name').text
        bndbox = obj.find('bndbox')
        xmin = bndbox.find('xmax').text
        xmax = bndbox.find('xmin').text
        ymin = bndbox.find('ymax').text
        ymax = bndbox.find('ymin').text
        parser.append([image_name, width, height, name,xmax,xmin,ymax,ymin])
        
    return parser
```

```python
# 타입 변환
cols = ['width', 'height', 'xmax', 'xmin', 'ymax', 'ymin']
df[cols] = df[cols].astype(float)
df[cols] = df[cols].astype(int)

df.info()
```

```python
# center x, center y
df['center x'] = ((df['xmax']+df['xmin'])/2)/df['width']
df['center y'] = ((df['ymax']+df['ymin'])/2)/df['height']
# w
df['w'] = (df['xmax']-df['xmin'])/df['width']
#h
df['h'] = (df['ymax']-df['ymin'])/df['height']
```

```python
#라벨링 인코딩
def label_encoding(x):
    labels = {'person':0, 'car':1, 'chair':2, 'bottle':3, 'pottedplant':4, 'bird':5, 'dog':6, 'sofa':7,
              'bicycle':8, 'horse':9, 'boat':10, 'motorbike':11, 'cat':12, 'tvmonitor':13, 'cow':14, 'sheep':15,
              'aeroplane':16, 'train':17, 'diningtable':18, 'bus':19}
    return labels[x]
```

```python
def save_data(filename, folder_path, group_obj):
    # 이미지 이동
    src = os.path.join('collect data', str(filename))
    dst = os.path.join(folder_path, str(filename))
    shutil.move(src, dst)  # 이미지 폴더로 이동
    
    # 라벨들 저장
    text_filename = os.path.join(folder_path, os.path.splitext(str(filename))[0] + '.txt')
    group_obj.get_group(filename).set_index('filename').to_csv(text_filename, sep=' ', index=False, header=False)
```
